```python
for i in range(len(data)):
    num=data.at[i,'위치'].find(',')
    if num != -1:
        data.at[i,'위치']=data.at[i,'위치'][:num].strip()
    else:
        pass
```

```python
for i in range(len(data)):
    tmp = gmaps.geocode(data['위치'][i], language='ko')
    
    if not tmp:
        continue
    else:
        tmp_loc = tmp[0].get("geometry")
        data['fac_lat'][i] = tmp_loc['location']['lat']
        data['fac_lng'][i] = tmp_loc['location']['lng']  
```
