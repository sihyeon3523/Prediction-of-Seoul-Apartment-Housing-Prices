# 8. 클러스터링

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("C:/Users/Yoo/Documents/Capstone/실거래(13~18).csv", thousands = ',')
```

```python
data = data.dropna().reset_index(drop=True)
data.head(3)
```

```python
data.rename(columns={'거래금액':'price', '건축년도':'con_year', '년':'year', '월':'month', '일':'day', '전용면적':'area', '층':'layer'}, inplace=True)
data = data.dropna().reset_index(drop=True)
```

```python
data =  data[['행정동','year','con_year','area','layer','land_price','D_hsch','D_msch','D_esch','D_sub','D_trans','D_hos','D_market',
          'D_library','D_factory','E_GDP','E_grow','E_CPI','E_rent','E_kospi','E_rate','A_benefit','A_pop_all','A_pop_65','A_pop_for',
          'A_par_all','A_par_apt','brand','Max_layer','A_top','A_bottom','price']]
```

```python
data.shape
```

```python
district_list = data.행정동.unique()
year_list = np.sort(data.year.unique())
district_list.shape, year_list.shape
```

```python
area_list = []

for i in district_list:
    A, B = np.percentile(data.loc[data['행정동'] == i, 'area'], [30, 70])
    area_list.append((A, B))
    
len(area_list)
```

```python
import itertools

dis_infor = []
dis_index = []

for i, j in itertools.product(district_list, year_list):
    A = np.where(data['행정동'].values == i)
    B = np.where(data['year'].values == j)
    C = np.intersect1d(A,B)
        
    dis_infor.append((i, j))
    dis_index.append(C)
```

```python
result_list = []

for n, (i, j) in enumerate(dis_infor):
    value1 = np.where(district_list == i) # 행정동 저장
    
    data1 = data.loc[dis_index[n]]
    X = data1.index
    X1 = data1[data1['area'] < area_list[value1[0][0]][0]].index
    X2 = data1[data1['area'] > area_list[value1[0][0]][1]].index
    X = X.difference(X1).difference(X2)
    
    result_list.append((i, j, X, X1, X2))
```

```python
for i in range(len(result_list)):
    print(len(result_list[i][2]), len(result_list[i][3]), len(result_list[i][4]))
```

```python
length = []
ddd = []

for i in range(len(result_list)):
    if len(result_list[i][2]) < 10:
        length.append(("{0} "+"{1}").format(result_list[i][0],result_list[i][1]))
    if len(result_list[i][3]) < 10:
        length.append(("{0} "+"{1}").format(result_list[i][0],result_list[i][1]))
    if len(result_list[i][4]) < 10:
        length.append(("{0} "+"{1}").format(result_list[i][0],result_list[i][1]))
        
for i in range(len(length)):
    ddd.append(length[i][:-4])

len(np.unique(np.array(ddd)))
```

```python
length = []

for i in range(len(result_list)):
    if len(result_list[i][2]) < 10:
        length.append("{0}".format(result_list[i][0]))
    if len(result_list[i][3]) < 10:
        length.append("{0}".format(result_list[i][0]))
    if len(result_list[i][4]) < 10:
        length.append("{0}".format(result_list[i][0]))

np.unique(np.array(length))
```

```python
# 최종 행정동 분류
prior = set(district_list)
poste = set(length)

district_select = prior.difference(poste)
len(district_select)
```

```python
# 최종 행정동만 남기기
final_list = []

for i in district_select:
    for j in range(len(result_list)):
        if result_list[j][0] == i:
            final_list.append(result_list[j])
```

```python
# 최종 데이터프레임 인덱스 만들기
col = []
LMS = ['_중', '_소', '_대']

for i, j in itertools.product(district_select, LMS):
    col.append("{0}{1}".format(i,j))
```

```python
final_data1 = pd.DataFrame(col, columns = ['area'])
final_data1.head(3)
```

```python
mean_list = []

for i in range(len(col)):
    for j in range(len(final_list)):
        if col[i][:-2] == final_list[j][0] and col[i][-1:] == '중':
            mean_list.append(round(data.loc[final_list[j][2],'price'].mean()))
        if col[i][:-2] == final_list[j][0] and col[i][-1:] == '소':
            mean_list.append(round(data.loc[final_list[j][3],'price'].mean()))
        if col[i][:-2] == final_list[j][0] and col[i][-1:] == '대':
            mean_list.append(round(data.loc[final_list[j][4],'price'].mean()))
```

```python
for i in range(len(col)):
    for j in range(len(final_list)):
        if col[i][:-2] == final_list[j][0] and col[i][-1:] == '중':
            print(i, j, final_list[j][0], final_list[j][1], col[i][-1:])
        if col[i][:-2] == final_list[j][0] and col[i][-1:] == '소':
            print(i, j, final_list[j][0], final_list[j][1], col[i][-1:])
        if col[i][:-2] == final_list[j][0] and col[i][-1:] == '대':
            print(i, j, final_list[j][0], final_list[j][1], col[i][-1:])
```
